JACARD SIMILARITY

1. Define a function lower_split which converts the whole text into lower case and splits phrases into words
2.  Define a Function Jaccard_Similarity and then returning the calculated similarity score using union and intersection
1. Using a For Loop and extracting the first 5 phrases from test Dataset
 >Extracting the required phrases

 >calling the lower_split and applying it on the phrase1 and phrase2

 >calling Jaccard_Similarity and finding the similarity between the 2 phrases

 >printing the 'phrase1' 'phrase2' and 'similarity score'

```python
import pandas as pd
df = pd.DataFrame(dataset['test'])
df.info()
```

```python
def lower(text):
    text = text.lower()
    return set(text.split())
def Jaccard_Similarity(doc1, doc2):
    intersection = doc1.intersection(doc2)
    union = doc1.union(doc2)
    return float(len(intersection)) / len(union)
for example in dataset['test'].select(range(999)):
    phrase1 = example['phrase1']
    phrase2 = example['phrase2']

    doc1 = lower(phrase1)
    doc2 = lower(phrase2)

    similarity = Jaccard_Similarity(doc1, doc2)
    print(f'phrase1:"{phrase1}"\t\tphrase2:"{phrase2}"\t\tscore:{similarity}')
```

```python
import gensim.downloader as api
import pandas as pd
model = api.load('word2vec-google-news-300')
```

```python
def lower(text):
    return text.lower()

p1 = dataset['test']['phrase1'][:999]
p2 = dataset['test']['phrase2'][:999]
similarities = []

for phrase1, phrase2 in zip(p1, p2):
    phrase1_lower = lower(phrase1)
    phrase2_lower = lower(phrase2)

    words1 = phrase1_lower.split()
    words2 = phrase2_lower.split()

    word_similarities = []
    for word1 in words1:
        for word2 in words2:
            if word1 in model.key_to_index and word2 in model.key_to_index:
                word_similarities.append(model.similarity(word1, word2))

    if word_similarities:
        phrase_similarity = sum(word_similarities) / len(word_similarities)
    else:
        phrase_similarity = 0

    similarities.append((phrase1, phrase2, phrase_similarity))

result = pd.DataFrame(similarities, columns=['phrase1', 'phrase2', 'similarity'])
print(result)

```

```python
import pandas as pd
import spacy
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, mean_squared_error
nlp = spacy.load('en_core_web_md')

p1 = dataset['test']['phrase1'][:100]
p2 = dataset['test']['phrase2'][:100]
labels = dataset['test']['label'][:100]

similarities = []

for sentence1, sentence2 in zip(p1, p2):
    doc1 = nlp(sentence1)
    doc2 = nlp(sentence2)
    similarity = doc1.similarity(doc2)
    similarities.append(similarity)

threshold = 0.8

predictions = [1 if sim >= threshold else 0 for sim in similarities]

conf_matrix = confusion_matrix(labels, predictions)
accuracy = accuracy_score(labels, predictions)
precision = precision_score(labels, predictions)
recall = recall_score(labels, predictions)
mse = mean_squared_error(labels, predictions)
print("Confusion Matrix:")
print(conf_matrix)
print("accuracy:", accuracy)
print("precision:", precision)
print("recall:", recall)
```

1.  The **SentenceTransformer** Class is *instantiated* with **all-MiniLM-L6-v2**

```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('all-MiniLM-L6-v2')
```


1.  Extracting the first 20 rows from 'phrase1' and 'phrase1' of test dataset
2.  Generating vector representations/embeddings of 'phrase1' and 'phrase2'into embedding1 and embedding2 using SentenceTransformer model
1. Then Calculating the cosine similarity between the vector representations/embeddings
2. Using a for loop iterating through the first 20 words and printing similarity scores


```python
phrases1 = dataset['test']['phrase1'][:20]
phrases2 = dataset['test']['phrase2'][:20]

embeddings1 = model.encode(phrases1, convert_to_tensor=True)
embeddings2 = model.encode(phrases2, convert_to_tensor=True)

cosine_score = util.cos_sim(embeddings1, embeddings2)

for i in range(20):
    cosine_score_value = cosine_score[i][i].item()
    print("{} \t\t{} \t\tScore:{:.4f}".format(phrases1[i], phrases2[i], cosine_score_value))

mean_cosine_similarity = cosine_score.diag().mean().item()
print("\nMean Cosine Similarity:", mean_cosine_similarity)
```
